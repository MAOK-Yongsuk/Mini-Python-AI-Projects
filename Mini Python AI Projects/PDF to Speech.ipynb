```python
import PyPDF4
from gtts import gTTS
from IPython.display import Audio

language =  'en'

read_file = 'testtext.pdf'
pdfReader = PyPDF4.PdfFileReader(read_file)

for pageNum in range(pdfReader.getNumPages()):
    pageObject = pdfReader.getPage(pageNum)
    text_to_speech = pageObject.extractText()
    
    gtts_object = gTTS(text = text_to_speech,
                      lang = language,
                      slow = False)
    
gtts_object.save("pdftospeechtest.mp3")

mp3_file = "pdftospeechtest.mp3"

Audio(mp3_file, autoplay = True)
    
```
